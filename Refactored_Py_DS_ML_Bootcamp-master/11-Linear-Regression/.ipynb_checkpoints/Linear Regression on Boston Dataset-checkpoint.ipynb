```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
print(boston.keys())
```

```python
print(boston['feature_names'])
```

```python
print(boston['target'])
```

```python
cols = boston['feature_names']

cols = np.append(cols, ['target'])


boston_df = pd.DataFrame(data= np.c_[boston['data'], boston['target']],
                     columns= cols)
```

```python
boston_df.head()
```

```python
boston_df.info()
```

```python
boston_df.describe()
```

```python
boston_df.columns
```

```python
sns.pairplot(boston_df[['CRIM', 'ZN', 'AGE', 'target']])
```

```python
sns.displot(boston_df['target']) #target is the price
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(boston_df.corr(), annot=True, linewidths=.5, ax=ax)
```

```python
X = boston_df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y = boston_df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train, y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
pd.DataFrame(lm.coef_, x_train.columns, columns=['Coeff'])
```

```python
#PREDITIONS
```

```python
predictions = lm.predict(x_test)
```

```python
plt.scatter(y_test, predictions)
```

```python
sns.displot((y_test-predictions))
```

```python
#EVALUATION
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test, predictions)
```

```python
mse = metrics.mean_squared_error(y_test, predictions)
print(mse)
```

```python
np.sqrt(mse)
```
